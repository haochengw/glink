```python
dbscan_data_path = "./dbscan_example_data.txt"
dbscan_result_path = "./dbscan_result.txt"
```

```python
id_cat_dict = {}
with open(dbscan_result_path, "r") as f:
    lines = f.readlines()
    for line in lines:
        items = line.replace("(", "").replace(")", "").split(",")
        id_cat_dict[int(items[1])] = int(items[0])
```

```python
merge_lines = []
data_file = open(dbscan_data_path, 'r')
merge_file = open("./dbscan_merge_result.txt", 'w', encoding='utf-8')

data_lines = data_file.readlines()
for line in data_lines:
    items = line.split(",")
    pid = int(items[0])
    merge_line = ""
    if pid in id_cat_dict:
        merge_line = line.strip() + "," + str(id_cat_dict[pid]) + "\n"
    else:
        merge_line = line.strip() + "," + "-1" + "\n"
    merge_file.write(merge_line)
```

```python
merge_file.close()
data_file.close()
```
