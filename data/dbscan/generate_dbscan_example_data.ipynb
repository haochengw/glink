```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1, _ = datasets.make_circles(n_samples=1000, factor=.6, noise=.02)
x1 = x1 / 4
x1[:, 0] = x1[:, 0] + 115
x1[:, 1] = x1[:, 1] + 35
```

```python
print("lat max: %f" % np.max(x1[:, 1]))
print("lat min: %f" % np.min(x1[:, 1]))
print("lat diff: %f" % (np.max(x1[:, 1]) - np.min(x1[:, 1])))
print()
print("lng max: %f" % np.max(x1[:, 0]))
print("lng min: %f" % np.min(x1[:, 0]))
print("lng diff: %f" % (np.max(x1[:, 0]) - np.min(x1[:, 0])))
```

```python
plt.scatter(x1[:, 0], x1[:, 1])
```

```python
with open("dbscan_example_data.txt", 'w', encoding='utf-8') as f:
    for i in range(len(x1)):
        f.write("%d,%f,%f,2020-05-30 12:00:00\n" % (i, x1[i][1], x1[i][0]))
```
